## Preprocessing dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../dataset/movies_metadata.csv")
df.info()
```

```python
df.columns
```

### Checking if null values are present in the dataset.

```python
df.isnull().sum()
```

### Checking how many duplicate rows are present

```python
df.duplicated().sum()
```

### Remove all duplicate rows

```python
df = df.drop_duplicates().reset_index(drop=True)
df.duplicated().sum()
```

### Select the following columns and generate a new dataframe from existing one.

+ Title
+ Overview
+ Genres
+ Tagline
+ Vote avg.
+ Popularity

```python
new_df = df[['title', 'overview', 'genres', 'tagline', 'vote_average', 'popularity']]
new_df.shape
```

### Check how many null value are present in the new dataframe.

```python
print("What columns are present in new dataframe: ", new_df.columns)
new_df.isnull().sum()
```

### Drop columns on the basis of title
**Because these the movies whose names are not present then how we'll recommend them.**

```python
print("Before: ", new_df.shape)
new_df = new_df.drop_duplicates(subset=['title'])
print("After: ", new_df.shape)
```

### Fill `overview` and `tagline` column with `''`.

```python
new_df['overview'] = new_df['overview'].fillna('')
new_df['tagline'] = new_df['tagline'].fillna('')
```

### How dataset looks like after all this?

```python
new_df.head(2).T
```

### we can clearly see that `genres` column contains a `dict` from which we've to extract the values

```python
new_df.info()
```

```python
import ast
```

```python
new_df['genres'] = new_df['genres'].apply(
    lambda X: " ".join([ obj['name'] for obj in ast.literal_eval(X) ])
)
```

### Finally, create a new column `tags` by combining 3 columns these are
+ overview
+ genres
+ tagline

```python
new_df["tags"] = new_df["overview"] + " " + new_df["genres"] + " " + new_df["tagline"]
new_df.shape
```

```python
new_df.info()
```

```python
new_df.to_csv("../dataset/updated_dataset.csv", index=False)
```
